```python
def check(users, trans):
    print('users.USER_ID is unique.') if users.USER_ID.nunique() == users.shape[0] else print("users.USER_ID isn't unique.")
    print()

    print('No missing values in users.') if users.count().min() == users.shape[0] else print('Missing values in users.')
    print('No missing values in trans.') if trans.count().min() == trans.shape[0] else print('Missing values in trans.')
    print()

    print('The users.AGE_CAT values {}.'.format(str(set(users.AGE_CAT.unique()))))
    print('The users.LOC_CAT values {}.'.format(str(set(users.LOC_CAT.unique()))))
    print('The users.INC_CAT values {}.'.format(str(set(users.INC_CAT.unique()))))
    print('The users.GEN values {}.'.format(str(set(users.GEN.unique()))))
    print()
    
    values = set()
    for col in users.filter(regex=("C201*")):
        values |= set(users[col].unique())
    print("The C* values {}.".format(values))

    values = set()
    for col in users.filter(regex=("W201*")):
        values |= set(users[col].unique())
    print("The W* values {}.".format(values))
    print()

    print('The trans.CHANNEL values {}.'.format(str(set(trans.CHANNEL.unique()))))
    print('The trans.TIME_CAT values {}.'.format(str(set(trans.TIME_CAT.unique()))))
    print('The trans.MC_CAT values {}.'.format(str(set(trans.MC_CAT.unique()))))
    print('The trans.CARD_CAT values {}.'.format(str(set(trans.CARD_CAT.unique()))))
    print('The trans.AMT_CAT values {}.'.format(str(set(trans.AMT_CAT.unique()))))
    print()

    data = trans.join(users[['USER_ID']].set_index('USER_ID'), on='USER_ID')
    print('All USER_ID in trans are correct.') if data.shape[0] == trans.shape[0] else print('Wrong USER_ID in trans.')
    
    data = trans.join(users[['USER_ID']].set_index('USER_ID'), on='USER_ID')
    print('All USER_ID in trans are correct.') if data.shape[0] == trans.shape[0] else print('Wrong USER_ID in trans.')
```

### Изменение исходный таблиц

```python
bank = bank_info.copy()
bank['POI_ID'] = bank['POI_ID'].astype(int)
# bank.set_index('POI_ID', inplace=True)

display(bank)
```

```python
import datetime

def fix_geo_coordinates(table, prefix=''):
    budapest_x = 650947
    budapest_y = 239670
    geo_scale = 250000
    table['%sGEO_X' % prefix] = table.eval('(%sGEO_X - @budapest_x) / @geo_scale' % prefix)
    table['%sGEO_Y' % prefix] = table.eval('(%sGEO_Y - @budapest_y) / @geo_scale' % prefix)

def init_users(_users):
    users = _users.copy()
    users['GEN'] = users['GEN'].apply(lambda g: 'm' if g == 1 else 'f')
    
#     fix_geo_coordinates(users, 'LOC_')
    
    for col in ('AGE_CAT', 'LOC_CAT', 'INC_CAT', 'GEN'):
        users[col] = users[col].astype('category')
        
    users['CARD_MONTHLY'] = ''
    users['WEALTH_MONTHLY'] = ''
    for col in users.filter(regex=("C201.0[1-6]")):
        users['CARD_MONTHLY'] += users[col].astype(str)
    for col in users.filter(regex=("W201.0[1-6]")):
        users['WEALTH_MONTHLY'] += users[col].astype(str)
    users.drop([
        col for col in users.columns
        if col.startswith('C201') or col.startswith('W201')
    ], axis=1, inplace=True)
    
    if 'TARGET_TASK_2' in users.columns:
        users['TARGET'] = users['TARGET_TASK_2'].apply(
             lambda dt: 1 if dt != '-' and dt >= "2014.07.01" else 0)
    return users
```

```python
def init_trans(_trans):
    trans = _trans.copy()

    for col in ('TIME_CAT', 'LOC_CAT', 'MC_CAT', 'CARD_CAT', 'AMT_CAT'):
        trans[col] = trans[col].astype('category')

    fix_float = lambda value: float(value) if value != '-' else None
    trans['GEO_X'] = trans['GEO_X'].map(fix_float)
    trans['GEO_Y'] = trans['GEO_Y'].map(fix_float)
    
#     trans = trans.join(bank, on='POI_ID', rsuffix='_')
#     sel = trans['GEO_X'].isnull() & trans['GEO_X_'].notnull()
#     trans.loc[sel, 'GEO_X'] = trans.loc[sel, 'GEO_X_']
#     sel = trans['GEO_Y'].isnull() & trans['GEO_Y_'].notnull()
#     trans.loc[sel, 'GEO_Y'] = trans.loc[sel, 'GEO_Y_']
    
#     fix_geo_coordinates(trans)
#     trans.drop(['GEO_X_', 'GEO_Y_'], axis=1, inplace=True)
   
    trans['DATE'] = trans['DATE'].astype('datetime64')
    return trans
```
